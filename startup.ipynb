```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns
import warnings
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
```

<h3><h>Data Exploration<hr>

```python
def plotter(dataframe, col1, col2, dep):
    plt.figure(figsize = (19,4))
    sns.set(style= 'darkgrid')

    plt.subplot(1,2,1)
    sns.regplot(data = dataframe, x=col1, y = dep, ci = None)
    plt.title(f'Relationship btw {col1} and {dep} is {dataframe[col1].corr(dataframe[dep]).round(2)}')

    plt.subplot(1,2,2)
    sns.regplot(data = dataframe, x=col2, y = dep, ci = None)
    plt.title(f'Relationship btw {col2} and {dep} is {dataframe[col2].corr(dataframe[dep]).round(2)}')
```

```python
plotter(data, 'R&D Spend', 'Administration', 'Profit')
plotter(data, 'Marketing Spend', 'Administration', 'Profit')
```

```python
df = data.copy()
```

```python
#using the describe method, we will ascertain the columns to be scaled by slecting columns with high std
cols_to_scale = []
for i in df.select_dtypes(include='number').columns:
    if df[i].describe()[2] >1500:
        cols_to_scale.append(i)
cols_to_scale
```

```python
#====================Scaling===============================
import joblib
scaler_holder = {}

for columns in cols_to_scale[:-1]:
    scaler = StandardScaler()
    df[columns] = scaler.fit_transform(df[[columns]])
    scaler_holder[columns + '_scaler'] = scaler
    joblib.dump(scaler, open(f'{columns}_scaler.pkl', 'wb'))
    print(f"{columns} successfully scaled")
```

```python
encoder = LabelEncoder()
df['State'] = encoder.fit_transform(df['State'])
joblib.dump(encoder, open('State_encoder.pkl', 'wb'))
df.head()
```

<h3><h>Modelling<h>

```python
x = df.drop('Profit', axis = 1)
y = df['Profit']

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.15)
```

```python
#Linear Regression model Creation

linreg = LinearRegression()
linreg.fit(xtrain, ytrain)

cv = linreg.predict(xtrain)
print(f"Cross Validation Score is {r2_score(cv,ytrain)}")
```

```python
#Model Testing

test_pred = linreg.predict(xtest)
print(f"Model Test Score is {r2_score(test_pred,ytest)}")
```
